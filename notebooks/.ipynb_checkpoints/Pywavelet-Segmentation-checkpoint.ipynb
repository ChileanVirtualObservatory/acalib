# Watershed Experiment
Image segmentation using Watershed.

Data: SDSS fits

## Loading Data

```python
hdulist = fits.open('../bindata/drC-crop.fits')
ndd = hdulist[0]
dataSdss = ndd.data

hdulist = fits.open('../bindata/drC-003630-r3-0083.fits')
ndd = hdulist[0]
dataSdssBig = ndd.data

hdulist = fits.open('../bindata/ALMA01000740.fits')
ndd = hdulist[0]
data = ndd.data
data = data[0]
subcube = data[slice(30,46), :,:]
dataAlma = np.sum(subcube,axis=0)

datas = [ (dataAlma, 'Alma integrated Cube'),(dataSdss, 'SDSS-Crop Dataset'), (dataSdssBig, 'SDSS Full Dataset') ]
```

## Thresholding

- Otsu Thresholding

- Morphological Transformation

- Watershed

```python
for data,title in datas:
    fig = plt.figure(figsize=(20,8))
    fig.suptitle(title,fontsize=14, fontweight='bold')
    #Loading Image
    img = np.array(data, dtype=np.uint8)
    ax = fig.add_subplot(131)
    ax.imshow(img)
    ax.set_title('Original')

    #Using otsu_thresh
    ret,thresh = cv2.threshold(img,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
    ax = fig.add_subplot(132)
    ax.imshow(thresh)
    ax.set_title('Otsu Thresholding')
        
    
    #Getting Labels
    markers = label(thresh)

    #Using Watershed Transform
    markers2 = watershed(thresh, markers, mask= img)
    ax = fig.add_subplot(133)
    ax.imshow(markers2)
    ax.set_title('Watershed')

    plt.tight_layout()
    plt.show()

    fts = regionprops(markers2)
    print("Features Found: %s" % len(fts))

    # noise removal
    kernel = np.ones((3,3),np.uint8)
    opening = cv2.morphologyEx(thresh,cv2.MORPH_OPEN,kernel, iterations = 2)

    markers = label(opening)

    markers2 = watershed(opening,markers, mask=img)    

    fig = plt.figure(figsize=(20,5))
    fig.suptitle(title+' Adding Morphological')

    ax = fig.add_subplot(141)
    ax.imshow(img)
    ax.set_title('Original')
    ax = fig.add_subplot(142)
    ax.imshow(thresh)
    ax.set_title('Otsu Thresholding')
    ax = fig.add_subplot(143)
    ax.imshow(opening)
    ax.set_title('Morphological Opening')
    ax = fig.add_subplot(144)
    ax.imshow(markers2)
    ax.set_title('Watershed Transform')
    plt.show()
    fts = regionprops(markers2)
    print("Features Found: %s" % len(fts))
    
```

# Threshold

First we make an automatic thresholding using OTSU Algorithm [https://en.wikipedia.org/wiki/Otsu%27s_method], the idea behing OTSU method is to find the threshold that maximizes the intra-class variance (Variance between foreground and background). Generally in astronomical data we have a lot of background and not so much foreground, so its a good threshold algorithm for this case.

In small images, we could lost a lot of objects, but in large images we need to use the opening morphological operator to get rid of the noise.

# Labeling

After Thresholding we label the different objects, looking for connected regions.

# Watershed

With the regions labeled, we use the watershed algorithm, that uses the grayscale image as topoghrapy and uses the idea of fountains and sinkholes to fill the spaces inside a labeled area (In the SDSS image we can see it clearly)

#Wavelet Transform

## Padding Data to make a square image

```python
fig = plt.figure(figsize=(20,8))
datas_pad = []
for i,(image, title) in enumerate(datas):
    xs = image.shape[0]
    ys = image.shape[1]
    lens = max(image.shape)
    if lens == xs:
        ys_lens = lens-ys
        ys_len = ys_lens/2
        if ys_lens%2 != 0:
            ys_pad = (ys_len,ys_len+1)
        else:
            ys_pad = (ys_len,ys_len)
        pad = ((0,0),ys_pad)
    else:
        xs_lens = lens-xs
        xs_len = xs_lens/2
        if xs_lens%2 != 0:
            xs_pad = (xs_len,xs_len+1)
        else:
            xs_pad = (xs_len,xs_len)
        pad = (xs_pad, (0,0))    
    
    image = np.array(image,dtype=np.uint8)
    data_pad = np.lib.pad(image,pad,'constant')
    datas_pad.append((data_pad, title))
    
    ax = fig.add_subplot(1,3,i+1)
    ax.imshow(data_pad)
    ax.set_title(title)
plt.show()
```

## Using 2D Stationary Wavelet Transform

```python
data_coeffs = []
for data, title in datas_pad:
    
    lens = np.max(data.shape)
    levels = pywt.swt_max_level(lens)
    coeffs = pywt.swt2(data_pad,'haar',level=levels)
    for LL, (LH,HL,HH) in coeffs:
        plt.imshow(HH)
        plt.plot()
```
