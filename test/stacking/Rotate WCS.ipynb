```python
import numpy as np 
import matplotlib.pyplot as plt
from astropy.io import fits
from astropy import wcs
import wcsaxes
%matplotlib inline
```

```python
path = 'fb2/M1-dss2b.fits'
hdulist =  fits.open(path, ignore_missing_end = True )

w = wcs.WCS(hdulist[0].header)
w.all_pix2world(30,40,0)
w
```

```python
fig = plt.figure()
fig.add_subplot(111, projection=w)
plt.imshow(hdulist[0].data, origin='lower', cmap='cubehelix')
plt.xlabel('RA')
plt.ylabel('Dec')
plt.show()
```
